```python
import pickle
import time

import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
def load_file(pkl_file_path):
    with open(pkl_file_path, 'rb') as file:
        loaded_data = pickle.load(file)
        return loaded_data
```

```python
hpo_time_1 = load_file('res/hpo_time_1.pkl')
hpo_time_2 = load_file('res/hpo_time_2.pkl')
hpo_time_4 = load_file('res/hpo_time_4.pkl')
```

```python
job_list = [1, 2, 4]
time_list = [hpo_time_1, hpo_time_2, hpo_time_4]
speedup_list = [time_list[0]/time for time in time_list]
```

```python
print(time_list)
print(speedup_list)
```

```python
def plot_time_and_speedup(num_list, time_list, speedup_list) -> None:
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(num_list, time_list, marker="o", label="Elapsed Time")
    plt.title("Elapsed Time")
    plt.xlabel("Number of jobs")
    plt.ylabel("Time (s)")
    plt.grid(True)
    plt.xticks(num_list, [int(n) for n in num_list])

    plt.subplot(1, 2, 2)
    plt.plot(num_list, speedup_list, marker="o", label="Speedup")
    plt.title("Speedup")
    plt.xlabel("Number of jobs")
    plt.ylabel("Speedup")
    plt.grid(True)
    plt.xticks(num_list, [int(n) for n in num_list])

    plt.tight_layout()
    plt.gcf().set_dpi(600)
    plt.savefig(f"res/hpo_jobs_effi_time_speedup.png", dpi=600)
    plt.show()
```

```python
plot_time_and_speedup(job_list, time_list, speedup_list)
```
